```python
# 2024 Term 1
import datetime
import csv

startDate = datetime.datetime(2023, 9, 4)
endDate = datetime.datetime(2023, 12, 8)
fileName = "Term1.csv"
schoolDays = []
weekNum = 1
TermNum = 1
header = ["Term", "Date", "Week", "Day", "Period","Class"]

def formatDate(dateToFormat):
    return (dateToFormat.strftime("%d-%m-%Y"))

class_mon = ["Y10_DT", "Y7_DL", "NC", "Y9_DL", "Y11_ICT", "NC"]
class_tue = ["NC", "Y7_DL", "Y8_STEAM", "Y10_CS","NC", "Y8_DL"]
class_wed = ["NC", "Y8_DL", "Y7_STEAM", "NC", "NC", "Y10_CS"]
class_thu = ["Y11_ICT","Y10_DT","Y9_STEAM","NC", "Y8_STEAM","Y9_DL"]
class_fri = ["NC", "Y11_ICT", "Y10_DT","Y9_STEAM","Y8_PSHE","Y7_STEAM"]



T1_holidays = [
    "28-09-2023",
    "16-10-2023",
    "17-10-2023",
    "18-10-2023",
    "19-10-2023",
    "20-10-2023",
    "13-11-2023",
]
with open(fileName, 'w', encoding='UTF8',newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    while( formatDate(startDate)!=formatDate(endDate+datetime.timedelta(days=1)) ):
        periodNum = 1
        if(not(startDate.strftime("%w")=="0" or startDate.strftime("%w")=="6") and formatDate(startDate) not in T1_holidays):
            if(startDate.strftime("%w")=="1"): # Monday
                for classes in class_mon:
                    writer.writerow([TermNum, formatDate(startDate), weekNum, startDate.strftime("%a"),"P"+str(periodNum), classes])
                    schoolDays.append(formatDate(startDate))
                    periodNum = periodNum + 1
            if(startDate.strftime("%w")=="2"): # Tuesday
                for classes in class_tue:
                    writer.writerow([TermNum, formatDate(startDate), weekNum, startDate.strftime("%a"),"P"+str(periodNum), classes])
                    schoolDays.append(formatDate(startDate))
                    periodNum = periodNum + 1
            if(startDate.strftime("%w")=="3"): # Wednesday
                for classes in class_wed:
                    writer.writerow([TermNum, formatDate(startDate), weekNum, startDate.strftime("%a"),"P"+str(periodNum), classes])
                    schoolDays.append(formatDate(startDate))
                    periodNum = periodNum + 1
            if(startDate.strftime("%w")=="4"): # Thursday
                for classes in class_thu:
                    writer.writerow([TermNum, formatDate(startDate), weekNum, startDate.strftime("%a"),"P"+str(periodNum), classes])
                    schoolDays.append(formatDate(startDate))
                    periodNum = periodNum + 1
            if(startDate.strftime("%w")=="5"): # Friday
                for classes in class_fri:
                    writer.writerow([TermNum, formatDate(startDate), weekNum, startDate.strftime("%a"),"P"+str(periodNum), classes])
                    schoolDays.append(formatDate(startDate))
                    periodNum = periodNum + 1
        if(startDate.strftime("%w")=="6"):
            weekNum = weekNum + 1
            
        startDate = startDate + datetime.timedelta(days=1)






for seq,schoolDay in enumerate(schoolDays):
    print(f"{seq+1}:{schoolDay}")

```
